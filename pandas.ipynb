```python
FakeDataCopySet = new_FakeDataFrame.copy()
### don't use assign operation like
# FakeDataCopySet = new_FakeDataFrame -> wrong (here it's a view of main DataFrame change of main will affect other.)
```

```python
## access value of data frame

# Don't use [i][j] Rather use loc to manipulate
# access the data
print(FakeDataCopySet.loc[2,3])

# change the data
FakeDataCopySet.loc[2,3] = 0.456789123

print(FakeDataCopySet.loc[2,3])
```

```python
FakeDataCopySet.columns
# we can also change just by assign a list of n elements here n is 10.
```

```python
FakeDataCopySet.index
```

```python
#if a row/column is not present here it'll create that in loc
FakeDataCopySet.loc[10,10] = 2.3
```

```python
FakeDataCopySet.head(11)
```

```python
FakeDataCopySet.drop(10,axis=1,inplace=True)
# axis=1 means column
# axis=0 means row
```

```python
FakeDataCopySet.loc[[1,2,3],[4,5]]
```

```python
FakeDataCopySet.loc[[i for i in range(200) if i%10==0],[5,6,7]]
```

```python
# : represent all
FakeDataCopySet.loc[[1,2,3],:]
```

```python
# making complex queries using loc
FakeDataCopySet.loc[((FakeDataCopySet[0]>0.9)&(FakeDataCopySet[2]>0.4))|(FakeDataCopySet[9]>0.95)]
```

```python
# Diff b/w loc and iloc

## loc uses name of columns whereas iloc uses index of column example is given below

'''
    1    2      3
A   0    1      2
B   3    4      5
C   6    7      8

df.loc['B',2] == df.iloc[1,2]

'''
```
